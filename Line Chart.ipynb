```python
import pandas as pd
import plotly.plotly as py

from ipywidgets import widgets 
from IPython.display import display
from plotly.graph_objs import *
from plotly.widgets import GraphWidget

# import data
complaints = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gantt_example.csv')

# we will define a function that will handle the input from the dropdown widget
def update_plot(complaint):
    #is_noise = complaints['Complaint Type'] == complaint['new']
    is_noise = complaints['Resource'] == complaint['new']
    temp = complaints[is_noise]
    data = temp.groupby(complaints['Borough'])['Resource'].count()
    
    x = []
    y = []
    for i in range(len(data)):
        x.append(data.index[i])
        y.append(data[i])

    graph.restyle({
                'x': [x],
                'y': [y],
            })
    graph.relayout({'title': 'Number of {} Complaints in New York by Borough'.format(complaint['new'])})

w = widgets.Dropdown(
    options= list(complaints['Resource'].unique()),
    value='A',
    description='Resource',  
)

graph = GraphWidget('https://plot.ly/~kevintest/1176/')

# observe will set a listener for activity on our dropdown menu
w.observe(update_plot, names="selected_label") 

display(w)
display(graph)
```

```python
us = pd.read_csv("./data/HPI_AT_us_and_census.csv")
state = pd.read_csv("./data/HPI_AT_state.csv")
county = pd.read_csv("./data/HPI_AT_BDL_county.csv")
```

```python
sf = county[county['County']=='San Francisco']
sf['HPI with 1980 base'] = sf['HPI'].apply(lambda x : round(float(x) / 219.61 * 100,2))
```

```python
county.head()
```

```python
test_state = "CA"
```

```python
all_county_since_1980 = county.query('State == "%s" and Year >= 1980' % test_state)

all_county_at_1980 = county.query('State == "%s" and Year == 1980' % test_state)[['County','HPI']]
all_county_at_1980.columns = ['County','HPI 1980']

merged = all_county_since_1980.merge(all_county_at_1980, how='inner', on='County')
merged['HPI 1980 base'] = np.round(merged['HPI'].astype(float)/merged['HPI 1980'].astype(float)*100,2)
county_names = sorted(list(set(merged['County'])))
```

```python
data = []
choices = []

for i,ct_name in enumerate(county_names):
    data.append(go.Scatter(
                x=merged.query('County == "%s"' % ct_name)['Year'],
                y=merged.query('County == "%s"' % ct_name)['HPI 1980 base'],
                name = ct_name,
                line = dict(color = 'orange'),
                opacity = 0.8))
    choices.append(dict(label = ct_name,
                 method = 'update',
                 args = [{'visible': [False]*i + [True] + [False]*(len(county_names)-i-1) + [True]*2}]))

    
trace_us = go.Scatter(
                x=us.query('Country == "USA" and Quarter == 1 and Year >= 1980')['Year'],
                y=us.query('Country == "USA" and Quarter == 1 and Year >= 1980')['HPI(1980 as base)'],
                name = "US",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

trace_ca = go.Scatter(
                x=state.query('State == "CA" and Quarter == 1 and Year >= 1980')['Year'],
                y=state.query('State == "CA" and Quarter == 1 and Year >= 1980')['HPI'],
                name = "CA",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)


data = data + [trace_us,trace_ca]


updatemenus = list([
    dict(active=-1,
         buttons=choices
    )
])

layout = dict(
    title = "Yearly HPI Comparison",
    xaxis = dict(range = ['1980','2017']),
    updatemenus=updatemenus
)

fig = dict(data=data, layout=layout)
iplot(fig, filename = "update_dropdown")
```

```python
all_county_since_1980 = county.query('Year >= 1980')

all_county_at_1980 = county.query('Year == 1980')[['State','County','HPI']]
all_county_at_1980.columns = ['State','County','HPI_1980']

merged = all_county_since_1980.merge(all_county_at_1980, how='inner', on=['State','County'])
to_drop = merged.query('HPI == "." or HPI_1980 == "."')['FIPS code'].drop_duplicates()
merged = merged[~merged['FIPS code'].isin(to_drop)]
merged['HPI 1980 base'] = np.round(merged['HPI'].astype(float)/merged['HPI_1980'].astype(float)*100,2)
state_names = sorted(list(set(merged['State'])))
combined_names = sorted(list(set(merged['State'] + ' - ' + merged['County'])))
```
